```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 15 09:58:43 2025

@author: mike
"""

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

# =====================================================
# Physical Constants and Stellar Parameters
# =====================================================
G       = 6.67e-11         # Gravitational constant (m^3 kg^-1 s^-2)
mu      = 8/13             # Mean molecular weight (dimensionless)
R_const = 8314.5           # Gas constant (J kg^-1 K^-1)
PI      = np.pi
kappa   = 0.02 * (1 + 0.7)   # Opacity (m^2 kg^-1), here assume hydrogen mass fraction X = 0.7

# Stellar parameters:
M_sun   = 1.989e30         # Solar mass (kg)
R_sun   = 6.9598e8         # Solar radius (m)
M_star  = 3.0 * M_sun      # Stellar mass for a 3 M_sun star (kg)
R_star  = 1.5 * R_sun      # Stellar radius (m)
Ms      = 0.5 * M_star     # Matching mass where outward and inward integrations are compared
m_c     = 1.0              # Starting mass for outward integration (to avoid singularity at m=0)#p_surface = (2.0/3.0) * G * M_star / (R_star**2 * kappa)


# =====================================================
# ODE Functions
# =====================================================
def lnpGrad(r, m, p):
    # Safeguard against division by zero in r
    r_safe = r if r > 1e-10 else 1e-10
    return -G * m / (4 * PI * (r_safe**4) * p)

def r3Grad(rho):
    # Safeguard against too-small density
    rho_safe = rho if rho > 1e-10 else 1e-10
    return 3.0 / (4 * PI * rho_safe)

def derivs(m, Y, K):
    """
    State vector Y = [r^3, ln(p)].
    Uses the polytropic relation: p = K * rho^(5/3) so that:
      rho = (p/K)^(3/5)
    Returns the derivatives [d(r^3)/dm, d(ln p)/dm].
    """
    r3, lnp = Y
    r = r3**(1.0/3.0) if r3 > 1e-10 else (1e-10)**(1.0/3.0)
    p = np.exp(lnp)
    rho = (p / K)**(3.0/5.0)
    dr3_dm = r3Grad(rho)
    dlnp_dm = lnpGrad(r, m, p)
    return [dr3_dm, dlnp_dm]

# =====================================================
# Integration Functions
# =====================================================
def outward_integration(K_local, p_c):
    """
    Outward integration from m = m_c to m = Ms.
    Starting conditions: r^3 = 0, ln(p) = ln(p_c)
    """
    init_out = [0.0, np.log(p_c)]
    m_grid = np.linspace(m_c, Ms, 500)
    sol = solve_ivp(lambda m, Y: derivs(m, Y, K_local),
                    t_span=(m_c, Ms), y0=init_out,
                    t_eval=m_grid, dense_output=True, max_step=1e27)
    return sol

def inward_integration(K_local):
    """
    Inward integration from m = M_star to m = Ms.
    Surface boundary conditions:
      r^3 = R_star^3, and p_surface determined via Eddington approximation:
        p_surface * kappa = (2/3)*G*M_star / R_star^2  =>  p_surface = (2/3)*G*M_star/(R_star^2*kappa)
    """
    r3_surface = R_star**3
    #p_surface = (2.0/3.0) * G * M_star / (R_star**2 * kappa)
    init_in = [r3_surface, np.log(p_surface)]
    m_grid = np.linspace(M_star, Ms, 500)  # integration backwards
    sol = solve_ivp(lambda m, Y: derivs(m, Y, K_local),
                    t_span=(M_star, Ms), y0=init_in,
                    t_eval=m_grid, dense_output=True, max_step=1e27)
    return sol

# =====================================================
# Grid Search Over Central Parameters
# =====================================================
# We now vary central pressure p_c and central temperature T_c over a wide range.
# The central density is given by: _c = (p_c * mu) / (R_const * T_c),
# and the polytropic constant is then: K = p_c / _c^(5/3).
p_surface = 1.8e6
T_s = 4700
# Use wider ranges:
p_c_values = np.linspace(1.4e15, 1.8e15, 10)  # central pressures 
T_c_values = np.linspace(1.3e7, 1.7e7, 10)      # central temperatures from 1e7 to 2e7 K

p_c_values = [3.6e15]
T_c_values = [1.75e7]
# We'll store data for each parameter pair in a list.
solutions_data = []  # list of dictionaries; each dict will store p_c, T_c, K_local, residual, and solutions

for p_c in p_c_values:
    for T_c in T_c_values:
        # Compute the central density from the ideal gas law:
        rho_c_local = (p_c * mu) / (R_const * T_c)
        # Polytropic constant:
        K_local = p_c / (rho_c_local**(5.0/3.0))
        #T_s = (mu / R_const) * (p_surface**(2/5)) * (K_local**(3/5))
        # Perform the outward integration from m = m_c to m = Ms.
        sol_out = outward_integration(K_local, p_c)
        if sol_out.status != 0:
            continue
        Y_match_out = sol_out.sol(Ms)
        r3_match_out = Y_match_out[0]
        lnp_match_out = Y_match_out[1]
        r_match_out = r3_match_out**(1.0/3.0)
        p_match_out = np.exp(lnp_match_out)
        
        # Perform the inward integration from m = M_star to m = Ms.
        rho_s_local = (p_surface * mu) / (R_const * T_s)
        K_local = p_surface / (rho_s_local**(5.0/3.0))
        sol_in = inward_integration(K_local)
        if sol_in.status != 0:
            continue
        Y_match_in = sol_in.sol(Ms)
        r3_match_in = Y_match_in[0]
        lnp_match_in = Y_match_in[1]
        r_match_in = r3_match_in**(1.0/3.0)
        p_match_in = np.exp(lnp_match_in)
        
        # Compute relative residuals:
        residual_r = abs(r_match_out - r_match_in) / ((r_match_out + r_match_in)/2)
        residual_p = abs(p_match_out - p_match_in) / ((p_match_out + p_match_in)/2)
        total_residual = (residual_r + residual_p) / 2.0
        
        solutions_data.append({
            'p_c': p_c,
            'T_c': T_c,
            'T_s': T_s,
            'K': K_local,
            'residual_r': residual_r,
            'residual_p': residual_p,
            'total_residual': total_residual,
            'sol_out': sol_out,
            'sol_in': sol_in
        })

# Sort the solutions by total_residual (ascending order)
solutions_data.sort(key=lambda d: d['total_residual'])

# Select the top 6 solutions (lowest mismatches)
top_solutions = solutions_data[:9]

print("Top 3 Solutions (lowest residuals):")
for idx, sol_data in enumerate(top_solutions):
    print(f"Solution {idx+1}: p_c = {sol_data['p_c']:.3e} Pa, T_c = {sol_data['T_c']:.3e} K, T_s = {sol_data['T_s']:.3e} K,"
          f"total residual = {sol_data['total_residual']:.3}")

# =====================================================
# Plotting the Integration Curves for the Top 6 Solutions
# =====================================================
plt.figure(figsize=(14, 12))

# Plot 1: Stellar Radius vs. Mass for Top 6 Solutions
plt.subplot(2, 1, 1)
for sol_data in top_solutions:
    p_c_val = sol_data['p_c']
    T_c_val = sol_data['T_c']
    T_s_val = sol_data['T_s']
    
    label_out = f"Outward: p_c={p_c_val:.2e}, T_c={T_c_val:.2e}"
    label_in  = f"Inward: p_c={p_c_val:.2e}, T_c={T_c_val:.2e}"
    
    sol_out = sol_data['sol_out']
    sol_in  = sol_data['sol_in']
    m_vals_out = sol_out.t
    r3_vals_out = sol_out.y[0]
    r_vals_out = r3_vals_out**(1.0/3.0)
    
    m_vals_in = sol_in.t
    r3_vals_in = sol_in.y[0]
    r_vals_in = r3_vals_in**(1.0/3.0)
    
    plt.plot(m_vals_out, r_vals_out, label=label_out)
    plt.plot(m_vals_in, r_vals_in, '--', label=label_in)
plt.axvline(Ms, color='k', linestyle=':', label="Matching mass, Ms")
plt.xlabel("Enclosed Mass m (kg)")
plt.ylabel("Radius r (m)")
plt.title("Stellar Radius vs. Mass for Best Solution and +/- 10%")
plt.legend(fontsize=8, loc='best')
plt.grid(True)

# Plot 2: Stellar Pressure vs. Mass for Top 6 Solutions
plt.subplot(2, 1, 2)
for sol_data in top_solutions:
    p_c_val = sol_data['p_c']
    T_c_val = sol_data['T_c']
    label_out = f"Outward: p_c={p_c_val:.2e}, T_c={T_c_val:.2e}"
    label_in  = f"Inward: p_c={p_c_val:.2e}, T_c={T_c_val:.2e}"
    
    sol_out = sol_data['sol_out']
    sol_in  = sol_data['sol_in']
    m_vals_out = sol_out.t
    lnp_vals_out = sol_out.y[1]
    p_vals_out = np.exp(lnp_vals_out)
    
    m_vals_in = sol_in.t
    lnp_vals_in = sol_in.y[1]
    p_vals_in = np.exp(lnp_vals_in)
    
    plt.plot(m_vals_out, p_vals_out, label=label_out)
    plt.plot(m_vals_in, p_vals_in, '--', label=label_in)
plt.axvline(Ms, color='k', linestyle=':', label="Matching mass, Ms")
plt.xlabel("Enclosed Mass m (kg)")
plt.ylabel("Pressure p (Pa)")
plt.title("Stellar Pressure vs. Mass for Best Solution and +/- 10%")
plt.legend(fontsize=8, loc='best')
plt.grid(True)
plt.savefig('Best Shooting Solutions.pdf',format='pdf')
plt.tight_layout()
plt.show()
```
